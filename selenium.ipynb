```python
# 安裝Selenium
# pip install selenium

# 載入需要的套件
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup as Soup
```

```python
# 開啟瀏覽器視窗(Chrome)
driver = webdriver.Chrome("/Users/dabaoku/Desktop/selenium/chromedriver")
```

```python
# 存要爬的data
df = pd.DataFrame(columns=['company_name', 'product_name', 'product_description', 'email'])
```

```python
# 先從臺灣雲市集>雲端方案>全部分類中爬出所有的“繼續閱讀”點下去的link
# https://www.tcloud.gov.tw
driver.get("https://www.tcloud.gov.tw/news_photo.aspx?n=113&sms=10165&page=1&PageSize=1000")
soup = Soup(driver.page_source, "lxml")
rawdata = soup.find_all(class_ ='read')
```

```python
# 根據所有link依序去爬資料
for data in rawdata:
    driver.get("https://www.tcloud.gov.tw" + data.find(class_='btn').get('href'))
    
    soup = Soup(driver.page_source, "lxml")
    box = soup.find(class_ ='box')

    company_name = box.find(class_='slogan').text
    product_name = box.find(class_='caption').text
    product_description = box.find(class_='p').text.replace('\n','')
    email = soup.find(title ='[另開新視窗]聯絡供應商').get('href').replace('mailto:','')
    
    df = df.append({'company_name': company_name,
                'product_name': product_name,
                'product_description': product_description,
                'email': email}, ignore_index=True)
    print(len(df))
```
